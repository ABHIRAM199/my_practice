## Regularization



Regularization is a cornerstone concept in machine learning. It serves as a strategy to mitigate overfitting by incorporating supplementary information into the model.

Overfitting occurs when a machine learning model exhibits superior performance on the training data but falters on unseen test data. In such a scenario, the model has essentially memorized the training data's idiosyncrasies and lacks the ability to generalize effectively to novel input examples. This often manifests as introducing noise into the output predictions when presented with unseen data.

Regularization techniques provide a remedy for overfitting. They are designed to:

* **Constrain Model Complexity:** Regularization methods aim to control the model's complexity, preventing it from becoming excessively intricate and adapting too closely to the training data's specific patterns.
* **Maintain Feature Inclusion:** Through regularization, all variables or features can potentially be retained within the model. However, the technique scales down the magnitude of these variables, striking a balance between model accuracy and generalization.
* **Reduce Coefficients:** Regularization typically operates by pushing the coefficients (weights) associated with features towards zero. In simpler terms, it reduces the influence of individual features without entirely eliminating them from the model.

Concisely, regularization techniques achieve their objective by decreasing the feature magnitudes while preserving the total number of features included in the model. This allows the model to maintain a degree of accuracy while enhancing its ability to generalize to unseen data.

### How does Regularization wroks?

**Regularization in Linear Regression: Cost Functions and the Impact of Complexity**

Regularization plays a crucial role in linear regression by introducing a penalty term to control the complexity of the model. Let's revisit the familiar linear regression equation:

y = β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ + b

where:

* y: The predicted value
* x₁, x₂, ..., xₙ: The features used for prediction
* β₀, β₁, ..., βₙ: The weights (or coefficients) associated with each feature
* b: The bias term of the model

Linear regression aims to optimize the β and b parameters to minimize a cost function. This cost function quantifies the model's overall performance by measuring the discrepancy between its predicted values and the actual target values (y).

**The Cost Function for Linear Regression: Residual Sum of Squares (RSS)**

In linear regression, the most common cost function is the Residual Sum of Squares (RSS). It's mathematically expressed as:

```
RSS = Σ(y_i - (β₀ + β₁x_i₁ + β₂x_i₂ + ... + βₙx_in + b))^2
```

where:

* Σ: Represents summation over all data points (i = 1 to n)
* y_i: The actual target value for data point i
* (β₀ + β₁x_i₁ + β₂x_i₂ + ... + βₙx_in + b): The predicted value for data point i based on the current model parameters

The RSS essentially calculates the squared difference between the predicted and actual values for each data point, and then sums these squared differences across all data points. By minimizing the RSS, we strive to find the β and b parameters that produce predictions closest to the actual target values.

The next passage will discuss how regularization is incorporated into the cost function to address overfitting.

## Techniques of Regularization

### L1 Regularization (Lasso Regression)

Lasso regression, also known as Least Absolute Shrinkage and Selection Operator (LASSO), constitutes another form of regularization for reducing model complexity. Similar to ridge regression, it introduces a penalty term into the cost function. However, the penalty term in Lasso regression employs the absolute values of the weights instead of their squares.

The cost function for Lasso regression is formulated as:

```
J(β₀, β₁, ..., βₙ) = RSS + λ Σ|βᵢ|
```

where:

* λ: The hyperparameter controlling the strength of the penalty term
* βᵢ: The weight (coefficient) associated with feature i

A key distinction between Ridge and Lasso regression lies in their treatment of coefficients. While Ridge regression shrinks coefficients towards zero, Lasso regression can actually drive them all the way to zero. This essentially eliminates certain features from the model altogether, performing feature selection alongside regularization.



### Ridge Regression (L2 Regularization)



Ridge regression is a technique that introduces a controlled amount of bias into the model to enhance its long-term prediction performance. It accomplishes this by reducing the model's complexity, hence the term "regularization."

Ridge regression modifies the cost function of linear regression by incorporating a penalty term. This penalty term, also known as the Ridge regression penalty, is calculated by multiplying a hyperparameter (λ) with the squared sum of the model's weights (coefficients).

The modified cost function for ridge regression is expressed as:

```
J(β₀, β₁, ..., βₙ) = RSS + λ Σ(βᵢ)^2
```

where:

* J(β₀, β₁, ..., βₙ): The modified cost function with the penalty term
* RSS: The original residual sum of squares cost function
* λ: The hyperparameter controlling the strength of the penalty term
* βᵢ: The weight (coefficient) associated with feature i

The penalty term serves to regularize the model's coefficients by pushing them towards zero. This effectively reduces the magnitude of the coefficients, leading to a less complex model.

Here's a key observation: as the value of λ approaches zero, the modified cost function progressively converges to the original cost function of linear regression. Consequently, for very small λ values, the model closely resembles a standard linear regression model.

Ridge regression is particularly beneficial in scenarios with highly correlated independent variables (collinearity). It can also address situations where the number of parameters (features) exceeds the number of data samples.


### Early Stopping

Early stopping, a regularization technique in machine learning, halts a model's training when validation loss ceases to improve. It guards against overfitting, applicable across supervised learning algorithms. By monitoring validation error during training, early stopping ceases training when validation error plateaus.

Its prime benefit lies in preventing overfitting and conserving resources by halting training when improvement ceases, saving time and costs by eliminating the need for extensive model iterations and hyperparameter tuning. Nevertheless, premature termination risks underfitting, potentially foregoing performance gains achievable with further training. Furthermore, optimal early stopping necessitates meticulous hyperparameter tuning, a laborious process.

### Dropouts



Dropout is a regularization method that approximates training a large number of neural networks with different architectures in parallel. As mentioned, it is preferrably used when training a large neural network.

*By dropping a unit out, we mean temporarily removing it from the network, along with all its incoming and outgoing connections*

Dropout has the effect of making the training process noisy, forcing nodes within a layer to probabilistically take on more or less responsibility for the inputs.
This conceptualization suggests that perhaps dropout breaks-up situations where network layers co-adapt to correct mistakes from prior layers, in turn making the model more robust.

It can be used with most types of layers, such as dense fully connected layers, convolutional layers, and recurrent layers such as the long short-term memory network layer. Dropout may be implemented on any or all hidden layers in the network as well as the visible or input layer.

`NOTE` : It is not used on the output layer.

````python
# Importing MNIST dataset from TensorFlow.keras.datasets
from tensorflow.keras.datasets import mnist

# Importing NumPy for numerical computations
import numpy as np

# Importing Matplotlib for data visualization
import matplotlib
import matplotlib.pyplot as plt

# Setting Matplotlib to display plots inline in the notebook
%matplotlib inline

# Loading the MNIST dataset into training and testing sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()
````

````python
# Selecting the first three images and their corresponding labels from the training dataset
images = X_train[:3]
labels = y_train[:3]

# Creating subplots for displaying multiple images
f, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 4))

# Iterating over the images and their corresponding axes for visualization
for index, (img, ax) in enumerate(zip(images, ax)):
    # Displaying the image on the axis
    ax.imshow(img, cmap='gray')
    # Removing axis labels for cleaner visualization
    ax.axis('off')
    # Adding text to show the label of the digit in the image
    ax.text(0.6, -2.0, f"Digit in the Image: {labels[index]}", size=15, ha="center")

# Displaying the plot
plt.show()
````

````python
# Importing the Sequential model from TensorFlow.keras.models
from tensorflow.keras.models import Sequential

# Importing layers such as Dense and Dropout from TensorFlow.keras.layers
from tensorflow.keras.layers import Dense, Dropout

# Importing optimizers such as SGD (Stochastic Gradient Descent) from TensorFlow.keras.optimizers
from tensorflow.keras.optimizers import SGD

# Importing TensorBoard callback for model visualization during training from TensorFlow.keras.callbacks
from tensorflow.keras.callbacks import TensorBoard

# Importing to_categorical utility function for one-hot encoding labels from TensorFlow.keras.utils
from tensorflow.keras.utils import to_categorical

# Importing EarlyStopping callback for stopping training early based on certain conditions from TensorFlow.keras.callbacks
from tensorflow.keras.callbacks import EarlyStopping

# Importing regularizers from TensorFlow.keras for applying regularization to the model
from tensorflow.keras import regularizers

# Importing datetime module for timestamping
from datetime import datetime
````

````python
###why did we reshaped ?? Question for you !

X_train = X_train.reshape(60000, 784)
X_test = X_test.reshape(10000, 784)
````

````python
###what is to_categorical doing here ? Question for you !

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
````

````python
# Define a Sequential model
model = Sequential()

# Add a fully connected layer with 10 units and ReLU activation function
# Input shape is (784,), corresponding to flattened images of size 28x28
model.add(Dense(10, activation='relu', input_shape=(784,)))

# Add a dropout layer with dropout rate of 0.2
model.add(Dropout(0.2))

# Add a fully connected layer with 256 units, ReLU activation function, and L2 regularization
model.add(Dense(256, activation='relu', kernel_regularizer=regularizers.l2()))

# Add a dropout layer with dropout rate of 0.2
model.add(Dropout(0.2))

# Add another fully connected layer with 256 units, ReLU activation function, and L2 regularization
model.add(Dense(256, activation='relu', kernel_regularizer=regularizers.l2()))

# Add a dropout layer with dropout rate of 0.2
model.add(Dropout(0.2))

# Add the output layer with 10 units and softmax activation function for multi-class classification
model.add(Dense(10, activation='softmax'))

# Compile the model
# Using categorical crossentropy as the loss function for multi-class classification
# Using Stochastic Gradient Descent optimizer with learning rate 0.001
# Evaluating the model based on accuracy
model.compile(loss='categorical_crossentropy', optimizer=SGD(lr=0.001), metrics=['accuracy'])
````

````python
# Creating a directory for TensorBoard logs with current timestamp
logdir = "logs/" + datetime.now().strftime("%Y%m%d-%H%M%S")

# Creating a TensorBoard callback to log training metrics for visualization
tensorboard_callback = TensorBoard(log_dir=logdir)

# Creating an EarlyStopping callback to monitor validation loss
# It will stop training if the validation loss doesn't decrease for 5 consecutive epochs
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=5)
````

````python
# model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
````

````python
# Training the model using the fit method
# X_train: Input training data
# y_train: Output training data (labels)
# batch_size=32: Batch size for training
# verbose=1: Show progress bar during training but suppress chatty output
# epochs=10: Number of epochs for training
# validation_data=(X_test, y_test): Validation data to evaluate the model performance during training
# callbacks=[tensorboard_callback, es]: List of callbacks to apply during training (TensorBoard for visualization and EarlyStopping for stopping training early)
training_history = model.fit(
    X_train,  # Input training data
    y_train,  # Output training data (labels)
    batch_size=32,
    verbose=1,  # Suppress chatty output; use Tensorboard instead
    epochs=10,
    validation_data=(X_test, y_test),
    callbacks=[tensorboard_callback, es],
)
````

##### Why not just use early stopping rather than regularisation ?



The main downside of early stopping is that this couples two tasks:

1. Algorithm to optimize the cost function j (eg gradient descent, adam etc)
2. Prevent overfitting (ie get more data, regularization)

because by stopping gradient decent early, we are sort of breaking whatever we are doing to optimize cost function J and simultaneously trying to not over fit. Rather than using early stopping, one alternative is just use L2 regularization then we can just train the neural network as long as possible(the downside here is we have to try a lot of values of the regularization parameter lambda and hence it becomes computationally expensive)
