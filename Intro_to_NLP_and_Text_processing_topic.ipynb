Natural Language Processing (NLP) is a field at the intersection of computer science, artificial intelligence, and linguistics. # **Text Processing Techniques**

####**Tokenization**
- Tokenization involves breaking down text into smaller units called tokens, which can be individual words or sentences.

- **Word Tokenization:** This process involves splitting a sentence into individual words. For example, the sentence "I love natural language processing" would be tokenized into ["I", "love", "natural", "language", "processing"].

- **Sentence Tokenization:** This process involves dividing a paragraph or text into individual sentences. For example, the paragraph "I love natural language processing. It's fascinating. I study it every day." would be tokenized into ["I love natural language processing.", "It's fascinating.", "I study it every day."].

####**Normalization**
Normalization is the process of converting text to a standard format, which makes it easier to process and analyze.

- **Lowercasing:** This involves converting all characters in the text to lowercase to ensure uniformity. For example, "Text Processing" becomes "text processing".
- **Stemming:** This involves reducing words to their root forms, which may not be a valid word. For example, "running" might be reduced to "run".
- **Lemmatization:** This involves reducing words to their base or dictionary form. Unlike stemming, lemmatization considers the context and converts the word to its meaningful base form. For example, "better" might be lemmatized to "good".
- **Removing Stop Words:** This involves eliminating common words that do not contribute much to the meaning of the text, such as "and", "the", "is", etc. Removing stop words helps in focusing on the important words in the text.


####**N-grams**
N-grams are contiguous sequences of 'n' items from a given text. They are used to capture context by analyzing combinations of words.

- **Unigrams:** Sequences of one word (e.g., "text", "processing").
- **Bigrams:** Sequences of two consecutive words (e.g., "text processing").
- **Trigrams:** Sequences of three consecutive words (e.g., "text processing techniques").

**Example of Applying Text Processing Techniques**
- **Consider the sentence:** "The quick brown fox jumps over the lazy dog."

**Word Tokenization:**
["The", "quick", "brown", "fox", "jumps", "over", "the", "lazy", "dog", "."]

**Normalization:**
- **Lowercasing:** ["the", "quick", "brown", "fox", "jumps", "over", "the", "lazy", "dog", "."]
- **Removing Stop Words:** ["quick", "brown", "fox", "jumps", "lazy", "dog", "."]
(Assuming stemming or lemmatization might not change these specific words significantly in this context)
N-grams:

- **Bigrams:** ["quick brown", "brown fox", "fox jumps", "jumps over", "over lazy", "lazy dog"]
- **Trigrams:** ["quick brown fox", "brown fox jumps", "fox jumps over", "jumps over lazy", "over lazy dog"]


**Applications of Text Processing Techniques**
- **Search Engines:** Tokenization and normalization help in matching user queries with relevant documents.
- **Text Classification:** Stop words removal and normalization improve the performance of text classifiers.
- **Sentiment Analysis:** Lemmatization helps in understanding the sentiment by analyzing the root form of words, N-grams help in understanding the context for better translation accuracy.

# **Text Representation**

**Bag of Words (BoW)**
- The Bag of Words model represents text as a collection of word counts, disregarding grammar and word order but keeping multiplicity.

- **How It Works:** Each document is represented as a vector where each dimension corresponds to a unique word in the corpus. The value in each dimension is the count of the word's occurrences in the document.

**Limitations:**
- **Ignores Word Order:** BoW treats text as a set of words without considering their order.
- **Ignores Context:** BoW does not capture semantic relationships between words.
TF-IDF (Term Frequency-Inverse Document Frequency)
TF-IDF is a statistical measure used to evaluate how important a word is to a document in a collection or corpus.

- **Term Frequency (TF):** Measures the frequency of a term in a document. It’s calculated as the number of times a term appears in a document divided by the total number of terms in that document.

- **Inverse Document Frequency (IDF):** Measures the importance of a term across all documents. It’s calculated as the logarithm of the total number of documents divided by the number of documents containing the term.


**TF-IDF Score:** The TF-IDF score is the product of TF and IDF.
**TF-IDF = TF * IDF**





####**Word Embeddings**
- Word embeddings are dense vector representations of words that capture their meanings, semantic relationships, and syntactic properties.

- **Word2Vec:** Generates word embeddings using two types of models:
- **Skip-gram:** Predicts the context words given a target word.
- **Continuous Bag of Words (CBOW):** Predicts the target word given the context words.
- **GloVe (Global Vectors for Word Representation):** Uses a co-occurrence matrix factorization approach to generate word embeddings. It considers the global word-word co-occurrence statistics from a corpus.
- **FastText:** Enhances Word2Vec by considering subword information. This means it breaks words into character n-grams and represents each word as the sum of its n-gram vectors, which helps in better handling of out-of-vocabulary words and capturing the internal structure of words.

Example of Applying Text Representation Techniques

**Consider the following three sentences:**

- "The cat sits on the mat."
- "The dog sits on the mat."
- "The cat lies on the rug."

**Bag of Words Representation**

- **Vocabulary:** ["The", "cat", "sits", "on", "the", "mat", "dog", "lies", "rug"]
- **Sentence 1 Vector:** [1, 1, 1, 1, 1, 1, 0, 0, 0]
- **Sentence 2 Vector:** [1, 0, 1, 1, 1, 1, 1, 0, 0]
- **Sentence 3 Vector:** [1, 1, 0, 1, 1, 0, 0, 1, 1]

**TF-IDF Representation**
- **TF Calculation:** For each term in each document, calculate the term frequency.
- **IDF Calculation:** Calculate the inverse document frequency for each term across all documents.
- **TF-IDF Score:** Multiply the TF by IDF for each term in each document.

**Word Embeddings (e.g., using Word2Vec)**
- **Training:** Train a Word2Vec model on a large corpus to learn vector representations for words.
- **Vector Representation:** Each word in the vocabulary is represented as a dense vector. For instance, "cat" might be represented as a vector [0.23, -0.15, 0.34, ...].

**Applications of Text Representation Techniques**
- **Document Similarity:** Using TF-IDF or word embeddings to compare the similarity between documents.
- **Text Classification:** Representing text using BoW, TF-IDF, or embeddings for training classifiers.
- **Information Retrieval:** Enhancing search engines by indexing documents with TF-IDF scores or word embeddings.
-  **Sentiment Analysis:** Utilizing word embeddings to capture the sentiment expressed in text.

####**Long Short-Term Memory (LSTM)**
- LSTMs are a type of RNN designed to overcome the vanishing gradient problem, making them capable of learning long-term dependencies.

**Key Components:**
- **Forget Gate:** Decides what information to discard from the cell state.
- **Input Gate:** Decides what new information to store in the cell state.
- **Output Gate:** Decides what part of the cell state to output.
- **Advantages:** LSTMs can remember information for long periods and are effective in tasks like language translation and speech recognition.


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM

# Create an LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(10, 1)))  # LSTM with 50 units, input shape (timesteps, features)
model.summary()
```

- GRUs are a simplified version of LSTMs, designed to achieve similar performance with fewer parameters.

**Key Components:**
- **Reset Gate:** Determines how much of the past information to forget.
- **Update Gate:** Determines how much of the new information to update to the current hidden state.
- **Advantages:** GRUs are computationally more efficient than LSTMs while maintaining similar effectiveness for many tasks.

####**Attention Mechanisms**
- Attention mechanisms allow models to focus on relevant parts of the input sequence, improving performance in tasks like machine translation and summarization.

**Key Features:**
- **Weighted Sum:** Calculates a weighted sum of the input elements, where the weights are learned.
- **Focus on Relevance:** Helps the model to focus on the most relevant parts of the input for a given task.


```python
import tensorflow as tf
from tensorflow.keras.layers import Layer, Input, Dense
from tensorflow.keras.models import Model
import numpy as np
import matplotlib.pyplot as plt

class Attention(Layer):
    def __init__(self, **kwargs):
        super(Attention, self).__init__(**kwargs)

    def build(self, input_shape):
        self.W = self.add_weight(name='attention_weight', shape=(input_shape[-1], input_shape[-1]), initializer='random_normal', trainable=True)
        self.b = self.add_weight(name='attention_bias', shape=(input_shape[-1],), initializer='zeros', trainable=True)
        super(Attention, self).build(input_shape)

    def call(self, inputs):
        e = tf.nn.tanh(tf.tensordot(inputs, self.W, axes=1) + self.b)
        a = tf.nn.softmax(e, axis=1)
        output = inputs * a
        return tf.reduce_sum(output, axis=1), a

# Create a simple model with attention
input_seq = Input(shape=(10, 64))
attention_layer, attention_weights = Attention()(input_seq)
output = Dense(1, activation='sigmoid')(attention_layer)
model = Model(inputs=input_seq, outputs=output)
model.summary()

# Create a model to get the attention weights
attention_model = Model(inputs=input_seq, outputs=attention_weights)

# Dummy data to visualize attention weights
dummy_data = np.random.random((1, 10, 64))

# Ensure predict function is traced only once
@tf.function
def get_attention_weights(input_data):
    return attention_model(input_data)

# Get the attention weights
attention_weights = get_attention_weights(dummy_data)

# Plot attention weights
plt.matshow(attention_weights.numpy().squeeze())
plt.colorbar()
plt.show()
```

####**Transformer Architecture**
- The Transformer model, based entirely on attention mechanisms, is designed to handle dependencies regardless of their distance in the input sequence.

**Key Components:**
- **Multi-Head Attention:** Allows the model to focus on different parts of the sequence simultaneously.
- **Positional Encoding:** Adds information about the position of words in the sequence, since the model itself has no inherent sense of order.
- **Feed-Forward Neural Networks:** Applied to each position separately and identically.
- **Layer Normalization:** Normalizes the output of the previous layer to stabilize and accelerate training.
- **Advantages:** Transformers handle long-range dependencies efficiently and have become the foundation for many state-of-the-art NLP models.


```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Embedding, MultiHeadAttention, LayerNormalization, Dropout
from tensorflow.keras.models import Model

def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    # Normalization and Attention
    x = LayerNormalization(epsilon=1e-6)(inputs)
    x = MultiHeadAttention(key_dim=head_size, num_heads=num_heads, dropout=dropout)(x, x)
    x = Dropout(dropout)(x)
    res = x + inputs

    # Feed Forward Part
    x = LayerNormalization(epsilon=1e-6)(res)
    x = Dense(ff_dim, activation="relu")(x)
    x = Dropout(dropout)(x)
    x = Dense(inputs.shape[-1])(x)
    return x + res

# Create a simple Transformer model
input_seq = Input(shape=(None, 128))
x = transformer_encoder(input_seq, head_size=128, num_heads=8, ff_dim=256, dropout=0.1)
output = Dense(10, activation='softmax')(x)

model = Model(inputs=input_seq, outputs=output)
model.summary()
```

####**BERT (Bidirectional Encoder Representations from Transformers)**
BERT is a pre-trained Transformer model designed to understand the context of a word based on its surroundings (bidirectional context).

- **Pre-training:** Trained on large corpora using tasks like Masked Language Modeling (MLM) and Next Sentence Prediction (NSP).
- **Fine-tuning:** Can be fine-tuned for specific NLP tasks such as question answering, sentiment analysis, and named entity recognition.
- **Advantages:** BERT achieves high performance by capturing deep bidirectional relationships in text.


```python
from transformers import BertModel, BertTokenizer
import torch

# Load pre-trained BERT model and tokenizer
model = BertModel.from_pretrained('bert-base-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Encode a sample text
inputs = tokenizer("Hello, my dog is cute", return_tensors='pt')
outputs = model(**inputs)

# Model summary
print(model)
```

####**PyTorch**
PyTorch is a deep learning framework that provides flexibility and speed for building and training models.

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Sample LSTM Model in PyTorch
class LSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim):
        super(LSTMClassifier, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        embedded = self.embedding(x)
        lstm_out, _ = self.lstm(embedded)
        out = self.fc(lstm_out[:, -1, :])
        return self.sigmoid(out)

# Model parameters
vocab_size = 10000
embedding_dim = 16
hidden_dim = 64
output_dim = 1

# Initialize model, loss function, and optimizer
model = LSTMClassifier(vocab_size, embedding_dim, hidden_dim, output_dim)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Model summary
print(model)
```
