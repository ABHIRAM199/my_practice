**Bias and variance trade off**



1.  *Bias* - difference b/w average prediction and actual value we are trying to predict.A model with high bias pays very little attention to the training data and oversimplifies(underfit) the model and leads to high error on training and test data
2.  *Variance* -spread of our data more variance implies more spread .High variance pays a lot of attention to training data(overfit) and does not generalise on the data which it hasn't seen before implies high error rates on test data

**Validation Techniques**

1)Basic Validation Split: Initially, the dataset is divided into training and test sets, typically in an 80-20 split. However, this approach can be problematic for smaller datasets, where insufficient training data may lead to biased models

2)Cross validation  :

Data Splitting: The dataset is divided into training and test sets, with a portion reserved for validation

Fold Creation: The training set is further divided into k folds.

Model Training and Evaluation: The model is trained k times, each time using a different combination of folds for training and validation.

Performance Assessment: Model performance metrics, such as R-squared, are computed for each fold.

Generalization Evaluation: Consistency in performance across folds indicates good generalization.

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits = 10,shuffle = True,random_state=1)

from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

degrees = 32 # no of data points

train_scores = []
test_scores = []

for degree in range(1,degrees):
  fold_train_scores = []
  fold_test_scores = []

  for train_index,test_index in kf.split(X):
    X_train,X_test = X[train_index],X[test_index]
    y_train,y_test = y[train_index],y[test_index]

    polyreg_scaled = make_pipeline(PolynomialFeatures(degree),scaler,LinearRegression())
    polyreg_scaled.fit(X_train,y_train)

    train_score = polyreg_scaled.score(X_train,y_train)
    test_score = polyreg_scaled.score(X_test,y_test)

    fold_train_scores.append(train_score)
    fold_test_scores.append(test_score)

  train_score = np.mean(fold_train_scores)
  test_score = np.mean(fold_test_scores)

  train_scores.append(train_score)
  test_scores.append(test_score)

plt.figure()
plt.plot(list(range(1,32)),train_scores,label = "train")
plt.plot(list(range(1,32)),test_scores,label = "test")
plt.legend(loc = "lower right")
plt.ylim(0,1)
plt.xlabel("degree")
plt.ylabel("R2-score")
plt.show()



```

The optimal degree of polynomial is the point where the testing score is maximized without a significant increase in the gap between training and testing scores.

**SIMULATION FOR REGULARISATION**

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet

#set a seed for reproducibility
np.random.seed(1)

#generate random data for x
X = np.random.rand(50,1)

#define a polynomial relationship b/w x and y with some noise

y = 0.7*(X**5)-\
    2.1*(X**4)+\
    2.3*(X**3)+\
    0.2*(X**2)+\
    0.3*(X)+\
    0.4*np.random.rand(50,1) #adding noise to the data

degree = 4

poly =PolynomialFeatures(degree) #create polynomial features to specified degree
X_poly = poly.fit_transform(X)   #transform X into polynomial features
X_poly_scaled = scaler.fit_transform(X_poly) #scale the polynomialfeatures


from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_poly_scaled,y,test_size = 0.2,random_state =1)

linear = LinearRegression()
linear.fit(X_train,y_train)
print("Linear model coefficients")
print(linear.coef_)

#Ridge Regression(L2 regression)
ridg = Ridge(alpha = 1)
ridg.fit(X_train,y_train)
print("Ridge regression coefficients")
print(ridg.coef_)

#Lasso Regression(L1 regression)
lass = Lasso(alpha=1)
lass.fit(X_train,y_train)
print("Lasso Regression coefficients")
print(lass.coef_)

#Elastic Net(L1+L2)
elastic = ElasticNet(alpha = 1)
elastic.fit(X_train,y_train)
print("ElasticNet regression coefficients")
print(elastic.coef_)

print("Train performance")
print("Linear R-square : ",linear.score(X_train,y_train))
print("Ridge R-square : ",ridg.score(X_train,y_train))
print("Lasso R-square : ",lass.score(X_train,y_train))
print("ElasticNet R-square : ",elastic.score(X_train,y_train))

print("Test performance")
print("Linear R-square : ",linear.score(X_test,y_test))
print("Ridge R-square : ",ridg.score(X_test,y_test))
print("Lasso R-square : ",lass.score(X_test,y_test))
print("ElasticNet R-square : ",elastic.score(X_test,y_test))


```
