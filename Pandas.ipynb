##Restructuring Data with Melt
Using **pd.melt**:

 * Method Application: Showed how to use pd.melt to restructure the DataFrame. This process involved merging multiple 'day' columns into a single 'day' column and creating a corresponding 'temp' column for temperatures.

 * Resulting Format: The outcome was a more streamlined DataFrame with consolidated day and temperature data.

Understanding **pd.melt** Parameters:
   * **id_vars:** Explained the use of id_vars to specify columns that should remain unchanged during the melt process.
   * **var_name and value_name**: Clarified the roles of var_name (to store the name of the melted columns) and value_name (to store the corresponding values).


```python
import pandas as pd
# Use the correct raw URL for the CSV file on GitHub
csv_url = 'https://raw.githubusercontent.com/yesdhruv/csv_files/master/weather.csv'


weather = pd.read_csv(csv_url)
# Display the DataFrame
weather
```

```python
# Perform the melt operation
weather_melt = pd.melt(weather,
                       id_vars=['MinTemp', 'MaxTemp', 'Rainfall'],  # Include actual columns in your DataFrame
                       var_name='day',
                       value_name='temp')
```

**.apply()**: This method is used to apply a function along the axis of a DataFrame or Series.

* Axis Specification in .apply():
 * When Axis=0: The function collapses the rows and applies mean_func to each column separately. This results in the mean of each column being calculated.

 * When Axis=1: The function collapses the columns and applies mean_func to each row. This results in the mean of the values in each row being calculated.


A custom function named **mean_func** is created to demonstrate the application of functions in Pandas. This function was designed to calculate the mean of a given set of values.

The function takes an input x and returns the mean calculated as the sum of x divided by the length of x.


```python
pd.Series([1,2,3,4,5])
```

```python
pd.DataFrame([[1,2,3,4],[5,6,7,8]],columns=['A','B','C','D'])
```

```python
pd.DataFrame({'A':[1,2,3,4,5],'B':[6,5,4,3,2]})
```

```python
import pandas as pd

def mean_func(x):
    """
    Calculate the mean of a given set of values.

    Parameters:
    - x: Input values for which the mean is calculated.

    Returns:
    - Mean value calculated as the sum of x divided by the length of x.
    """
    # Convert the values to numeric to handle mixed data types
    x_numeric = pd.to_numeric(x, errors='coerce')
    return x_numeric.sum() / len(x_numeric)

# Creating a sample DataFrame
data = {'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
        'Min_Temperature': [18, 20, 22, 19, 17],
        'Max_Temperature': [25, 28, 24, 26, 23]}
df = pd.DataFrame(data)
df
```

```python
43/3
```

```python
df.apply(mean_func,axis=1)
```

```python
# Applying mean_func with axis=0 (Column-wise mean)
column_means = df.apply(mean_func, axis=0)


# Displaying the results
print("Column-wise Mean:")
print(column_means)
```

```python
# Applying mean_func with axis=1 (Row-wise mean)
row_means = df.apply(mean_func, axis=1)

print("\nRow-wise Mean:")
print(row_means)
```

## Strategies for Handling Missing Data

Two primary strategies for handling missing data are deletion and imputation.





### Deletion Methods

* Listwise Deletion
 * Involves removing entire rows from a dataset if any value in the row is missing. This is a straightforward approach but can lead to significant data loss.

* Pairwise Deletion
 * Only the specific missing data point is omitted, not the entire row. Useful in statistical analyses like correlation, utilizing the remaining available data.




```python
# Import NumPy and Pandas
import numpy as np
import pandas as pd

# Create a sample DataFrame with missing values
data = {
    'Name': ['Alice', 'Bob', 'Ali', 'David', 'Eva'],
    'Age': [25, 30, np.nan, 22, 28],
    'Salary': [50000, np.nan, 60000, 400,np.nan],
    'Department': ['HR', 'IT', 'Finance', np.nan, 'Marketing']
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)
df
```

```python
# Deletion: Listwise Deletion
df_listwise = df.dropna()
print("\nDataFrame after Listwise Deletion:")
df_listwise
# print(df_listwise)
```

```python
# Deletion: Pairwise Deletion
df_pairwise = df.dropna(subset=['Name', 'Salary'])
print("\nDataFrame after Pairwise Deletion:")
df_pairwise
```

```python
df.fillna(9)
```

 Pros and Cons of Deletion
- Straightforward and easy to implement.
- Can introduce bias, affecting the representativeness and reliability of the dataset.
- Best suited for scenarios with minimal missing data or when the missing data is randomly distributed.

#### Pros and Cons of Imputation
- Helps in preserving the original size of the dataset.
- Can introduce inaccuracies or biases if not implemented with an understanding of the dataset's characteristics.
- Requires careful consideration of the nature of the missing data and the overall dataset structure.
- Often preferred over deletion methods in scenarios where retaining the maximum amount of data is crucial.

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# Sample DataFrame with missing values
data = {'A': [1, 2, np.nan, 4, 5],
        'B': [10, 20, 30, np.nan, 50],
        'C': [100, 200, 300, 400, 500]}

df = pd.DataFrame(data)
```

```python
# Check for missing values in the DataFrame
if df.isnull().values.any():
    # Handling missing values by filling with the mean
    df_filled = df.fillna(df.mean())

    # Check if there are still any missing values
    if df_filled.isnull().values.any():
        raise ValueError("DataFrame still contains missing values. Check and handle missing values appropriately.")
else:
    df_filled = df.copy()  # No missing values, create a copy of the original DataFrame
```

```python
# Check if the DataFrame contains numeric data
numeric_columns = df_filled.select_dtypes(include=[np.number]).columns
if len(numeric_columns) == 0:
    raise ValueError("DataFrame does not contain numeric columns. KNN imputation requires numeric data.")

# Imputation: K-Nearest Neighbors (KNN) Imputation
imputer = KNNImputer(n_neighbors=2)
df_knn_imputed = pd.DataFrame(imputer.fit_transform(df_filled), columns=df_filled.columns)
```

```python
# Displaying Results
print("\nOriginal DataFrame:")
print(df)

print("\nDataFrame after Handling Missing Values:")
print(df_filled)

print("\nDataFrame after KNN Imputation:")
print(df_knn_imputed)
```
