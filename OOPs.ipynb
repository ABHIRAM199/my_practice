### Dunder Methods: String Representation of Objects

The `__str__` method in Python is a special method used to provide a meaningful string representation of an object.

***When you use the built-in `str()` function or `print()` function on an object, Python invokes the `__str__` method if it is defined for that object.***

- **Purpose:**
  - Defines the human-readable string representation of an object.
  - Intended for display purposes or creating meaningful output for end-users.

- **Naming Convention:**
  - The method name is double-underscored (`__str__`), indicating its special status.

- **Syntax:**
  - ```python
    def __str__(self):
        # return a string representation of the object
    ```

- **Usage:**
  - When you use `str(obj)` or `print(obj)`, Python looks for the `__str__` method in the object's class to obtain the string representation.

- **Fallback to `__repr__`:**
  - If `__str__` is not defined, Python falls back to the `__repr__` method for obtaining a string representation.

````python
class Car:
    def __init__(self, name, mileage):
        self.name = name
        self.mileage = mileage
````

## Inheritance

inheritance.png

**Key Terms**

1. **Inheritance:** A mechanism in object-oriented programming where a new class (subclass or derived class) inherits attributes and behaviors from an existing class (superclass or base class).
2. **Superclass:** The existing class from which attributes and methods are inherited by a new class. Also known as a base class or parent class.
3. **Subclass:** The new class that inherits attributes and methods from an existing class. Also known as a derived class or child class.
4. **Base Class:** Another term for a superclass, indicating the class from which other classes inherit.
5. **Derived Class:** Another term for a subclass, indicating the class that inherits from another class.

The syntax for inheritance in Python involves creating a subclass that inherits from a superclass.

```python
class Superclass:
    # Superclass attributes and methods

class Subclass(Superclass):
    # Subclass attributes and methods
```

Explanation of the syntax:

- `class Superclass:`: This line defines the superclass.
- `class Subclass(Superclass):`: This line defines the subclass, and the `(Superclass)` indicates that the subclass inherits from the superclass.

````python
class Animal:
    def __init__(self, species):
        self.species = species

    def eat(self):
        print("I am eating!!!")
````

````python
class Cat(Animal):
    def __init__(self, name):
        # To set specific data for the Cat, we need to call the Animal initializer
        # Using Animal.__init__(self, "Cat") invokes the constructor of the Animal superclass
        # This ensures that the 'species' attribute of the Cat is initialized to "Cat"
        Animal.__init__(self, "Cat")

        # super().__init__("Cat")

        # Now, we can set the 'name' attribute specific to the Cat subclass
        self.name = name

    def speak(self):
        print("Meoooewww Meooow!!!!")
````

````python
class Dog(Animal):
    def __init__(self, name):
        Animal.__init__(self, "Dog")
        self.name = name

    def speak(self):
        print("Boooow Bhhoowww!!!!")
````

### `super()` Function

In Python, `super()` is a built-in function that is used to call a method from a superclass.

- **Method Resolution Order (MRO):**
  - In Python, when a method is called, the interpreter searches for it in the current class. If not found, it looks in the superclass, then in the superclass of the superclass, and so on. This order is known as the Method Resolution Order (MRO).
  - `super()` helps maintain the MRO by allowing a subclass to invoke a method from its superclass.

**Syntax:**
```python
class Subclass(Superclass):
    def some_method(self):
        super().some_method()  # Call the same method from the superclass
```

- `super()` returns a temporary object of the superclass, allowing you to call its methods.
- It is often used inside a method in a subclass to call the corresponding method in the superclass.
- `super()` without arguments is equivalent to `super(__class__, self)` where `self` is the instance of the subclass.

````python
class Animal:
    def __init__(self, species):
        self.species = species
        print('self inside Animal class:')
        print(self)

    def eat(self):
        print("I am eating!!!")
````

````python
Animal-- superClass

Animal.__init__(self,"Cat")

super().__init__("Cat")
````

````python
class Cat(Animal): # syntax to inherit
    def __init__(self, name):
        print('self inside Cat class:')
        print(self)
        super().__init__("Cat") # the object is passed automatically
        self.name = name

    def speak(self):
        print("Meoooewww Meooow!!!!")
````

````python
c = Cat("Tom")
print(c.name)
````

### Public Variables



````python
class Car:
    def __init__(self, make, model, year):
        # Public variables
        self.make = make
        self.model = model
        self.year = year

    def display_info(self):
        print(f"{self.year} {self.make} {self.model}")

# Creating an instance of the Car class
my_car = Car("Toyota", "Camry", 2022)

# Accessing and modifying public variables
print(f"Make: {my_car.make}")
print(f"Model: {my_car.model}")
print(f"Year: {my_car.year}")

# Modifying a public variable
my_car.year = 2023

# Displaying updated information
my_car.display_info()
````

* In Python, variables without any underscore prefix are considered public by convention.

* In this example, `make`, `model`, and `year` are public variables. They are accessible from outside the class, and their values can be easily modified. The `display_info` method demonstrates how to use these public variables to display information about the car.

### Protected Variables

````python
class MyClass:
    def __init__(self):
        self._protected_var = 10
        self.test=100
        self.__test2=1000

    def get_protected_var(self):
        return self._protected_var

    def set_protected_var(self, value):
        if value > 0:
            self._protected_var = value
        else:
            print("Value must be greater than 0.")

# Creating an instance of MyClass
obj = MyClass()
````

````python
obj.__test2
````

````python
obj.test
````

````python
obj._protected_var
````

````python


# Accessing the protected variable using a getter method
print("Initial Value:", obj.get_protected_var())

# Modifying the protected variable using a setter method
obj.set_protected_var(20)
print("Modified Value:", obj.get_protected_var())

# Trying to access the protected variable directly (not recommended)
print("Direct Access:", obj._protected_var)
````

* In Python, a protected variable is conventionally denoted by a single leading underscore (`_`).

* While it doesn't provide strict access control like private variables (those with a double leading underscore), it signals that the variable is intended for internal use and should not be accessed directly from outside the class.

- `_protected_var` is a protected variable within the class `MyClass`.
- The `get_protected_var` method is a getter method to access the protected variable.
- The `set_protected_var` method is a setter method to modify the protected variable, but with validation (value must be greater than 0).
- Direct access to `_protected_var` from outside the class is discouraged, but it's still possible. The convention is to use the getter and setter methods.

* **Security**: Sensitive data and operations are hidden from the user.

* **Modularity**: Changes in the abstracted part do not affect the user's interaction with it.

Two ways to implement Abstraction:-

#### 1. Abstract Class

- Key Points:
 - Abstract class "Shape" with an abstract method calculate_area().
 - Rectangle and Circle are subclasses providing their implementations.





````python
# Demonstrates polymorphism.
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        return 100

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return 3.14 * self.radius * self.radius
````

## Polymorphism

- **Key terms**
 * Polymorphism, derived from the Greek words "poly" (many) and "morph" (shape).

 * Polymorphism allows objects of different classes to be treated as objects of a common superclass. It's essential for writing flexible and reusable code.

There are mainly two types of Polymorphism:

### 1. Method Overriding

Also known as Run-Time Polymorphism

Subclasses provide a specific implementation of a method that is already defined in its superclass.

Example:

````python
class Bird:
    def fly(self):
        return "Some birds can fly"

class Sparrow(Bird):
    def fly(self):
        return "Sparrow flies"

class Ostrich(Bird):
    def fly(self):
        return "Ostrich cannot fly"

# Usage
sparrow = Sparrow()
ostrich = Ostrich()
print(sparrow.fly())  # Output: Sparrow flies
print(ostrich.fly())  # Output: Ostrich cannot fly
````

#### What is Duck Typing?
The famous saying, "If it looks like a duck and quacks like a duck, then it probably is a duck," encapsulates the essence of Duck Typing

* Duck Typing is a form of run-time polymorphism.
* **Flexible Type Checking:** Unlike traditional polymorphism, which relies on inheritance hierarchies and interfaces,  Duck Typing allows for more flexible type checking.

````python
class Bird:
    def fly(self):
        return "This bird can fly"

class Airplane:
    def fly(self):
        return "This airplane can fly"

def test_flying_ability(flying_object):
    print(flying_object.fly())

# Usage
bird = Bird()
airplane = Airplane()
test_flying_ability(bird)        # Output: This bird can fly
test_flying_ability(airplane)    # Output: This airplane can fly
````

### 2. Method Overloading
Also known as Compile Time Polymorphism or Parameter Polymorphism

 #### Using Default Parameter Values

1. Default Parameter Values:

 * Define default values for function parameters.
 * Allows a function to be called with fewer arguments than it is defined to accept.
2. Optional Parameters:

 * Parameters with default values become optional.
 * Provide flexibility in function calls.

````python
class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c

# Example Usage:
calc = Calculator()

# Example 1: Adding two numbers
result1 = calc.add(5, 3)
print(f"Result 1: {result1}")

# Example 2: Adding three numbers
result2 = calc.add(2, 4, 6)
print(f"Result 2: {result2}")
````
